```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("dataset.csv")
```

Count number of startups in each sector

```python
sector_counts=df['Sector'].value_counts()
print(sector_counts)
```

Top 10 sector by highest number of startups

```python
sector_counts=sector_counts.head(10)
```

```python
plt.figure(figsize=(10,6))
sector_counts.plot(kind=('bar'))
plt.xlabel('Sector')
plt.ylabel('Number of unicorns')
plt.title('TOP 10 SECTORS WITH  HIGHEST UNICORNS')
plt.xticks(rotation=50)
plt.tight_layout()
plt.show()
```

```python
df['Year']=pd.to_datetime(df['Entry'],format='%b/%Y').dt.year

yearly_counts = df['Year'].value_counts().sort_index()
```

```python
plt.figure()
sns.barplot(x=yearly_counts.index, y=yearly_counts.values)
plt.xlabel('Year')
plt.ylabel('Number of Unicorns Started')
plt.title('Unicorns Started Each Year')

plt.xticks(rotation=45)
plt.show()
```
